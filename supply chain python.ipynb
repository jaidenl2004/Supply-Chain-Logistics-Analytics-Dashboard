```python
import pandas as pd
df = pd.read_csv ( "/content/logistics data.csv")
```

```python
df.info
df.isna().sum()
df.describe
```

```python
df['Shipment_Date']=pd.to_datetime(df['Shipment_Date'], errors='coerce')
df['Delivery_Date']=pd.to_datetime(df['Delivery_Date'],errors='coerce')
```

```python
df['Actual Transit days'] =(df['Delivery_Date']-df['Shipment_Date']).dt.days
df['Delayed Flag'] =df['Status'].apply(lambda x: 1 if x == 'Delayed' else 0)
df['Cost per mille'] =df['Cost']/df['Distance_miles']
df['Cost per Kg'] =df['Cost']/df['Weight_kg']
```

```python
# Delivery speed efficiency (lower is better)
df['Cost Efficiency'] = df['Cost'] / df['Actual Transit days']

# Weight category
df['Weight Category'] = pd.cut(
    df['Weight_kg'],
    bins=[0, 5, 20, 50, 200],
    labels=['Light', 'Medium', 'Heavy', 'Very Heavy']
)

# Distance category
df['Distance Category'] = pd.cut(
    df['Distance_miles'],
    bins=[0, 300, 1000, 2000, 5000],
    labels=['Local', 'Regional', 'Cross-country', 'International']
)

# Carrier performance flag
```

```python
df.to_csv("clean_logistics_data.csv", index=False)
```
